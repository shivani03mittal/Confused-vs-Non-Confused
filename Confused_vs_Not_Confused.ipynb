```python
X = df.drop('Label',axis=1)
Y = df['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.70, random_state=500)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# 10 Fold Cross validation

```python
from sklearn.model_selection import KFold
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

# Applying Different classification Algorithms

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.pipeline import Pipeline
from  imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_val_score

```

# Logistic Regression

```python
steps = [('over', SMOTE()), ('model', LogisticRegression())]
pipeline = Pipeline(steps=steps)
pipeline.fit(X_train,Y_train)
```

```python
pred= pipeline.predict(X_test)
pred
```

```python
score = cross_val_score(pipeline, X_test, Y_test, scoring='accuracy', cv=cv, n_jobs=-1)
precision = cross_val_score(pipeline, X_test, Y_test, scoring='precision', cv=cv, n_jobs=-1)
roc_auc = cross_val_score(pipeline, X_test, Y_test, scoring='roc_auc', cv=cv, n_jobs=-1)

print("min and max accuracy:",np.min(score),',',np.max(score))
print("Mean accuracy of model:",np.mean(precision))
print("")
print("min and max precision:",np.min(precision),',',np.max(precision))
print("Mean precision of model:",np.mean(precision))
print("")
print("min and max roc_auc:",np.min(roc_auc),',',np.max(roc_auc))
print("Mean area under roc curve of model:",np.mean(roc_auc))
```

```python
print("Standard deviation of accuracy :",np.std(score))
print("Standard deviation of area under roc :",np.std(roc_auc))
print("Standard deviation of precision :",np.std(precision))
```

```python
def plot_confusion_matrix(Y_test,pred):
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(Y_test, pred)

    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax, fmt='g')

    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score 
from sklearn.metrics import  precision_recall_fscore_support
def stats(Y_test,pred):
        score = precision_recall_fscore_support(Y_test,pred)
        
        print("Accuracy of model is :",accuracy_score(Y_test,pred))
        print("Area under ROC curev is :",roc_auc_score(Y_test,pred))
        print('precision: ', score[0])
        print('recall: ', score[1])
        print('f1_score: ', score[2])
        print('support: ', score[3])
        
        plot_confusion_matrix(Y_test,pred)
        
```

```python
stats(Y_test,pred)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
model = Sequential()
model.add(Dense(df.shape[1],activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])#loss='mse'mean squared error
history = model.fit(X_train, Y_train, validation_split=0.3, epochs=150)
print(history.history.keys())
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
plt.plot(history.history['val_accuracy'])
plt.plot(history.history['val_loss'])
plt.legend(['accuracy', 'loss'], loc='center right')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.show()
```
